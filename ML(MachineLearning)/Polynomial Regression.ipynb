```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
# create a dataset 
X = 6 * np.random.rand(100,1)-3 # -3 are some outliers
Y = 0.5 * X**2 + 1.5*X + 2 + np.random.randn(100,1) # quadratic equation used y = 0.5 * X**2 + 1.5*X + 2+outliers
plt.scatter(X,Y,color = 'green')
plt.xlabel("X DataSet")
plt.ylabel("Y DataSet")
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
```

```python
regression.fit(X_train,Y_train)
```

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(Y_test,regression.predict(X_test))
score 
```

```python
plt.plot(X_train,regression.predict(X_train),color='r')
plt.scatter(X_train,Y_train,color='b')
```

```python
# let's solve the above above problem(Best fit line) with polymonial regression
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2,include_bias=True)
```

```python
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)
```

```python
X_train_poly
```

```python
X_test_poly
```

```python
# now again check the score 
regression.fit(X_train_poly,Y_train)
Y_pred = regression.predict(X_test_poly)
score = r2_score(Y_test,Y_pred)
print(score)
```

```python
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
# Now check the best fit line which will be quit amazing
plt.scatter(X_train,regression.predict(X_train_poly))
plt.scatter(X_train,Y_train)
```

```python
# Now prediction for new data
X_new = np.linspace(-3,3,200).reshape(200,1)
X_new_poly = poly.transform(X_new)
X_new_poly
```

```python
# now let's do the predication 
Y_new = regression.predict(X_new_poly)
plt.plot(X_new,Y_new,"r-",linewidth=2,label="Predications")
plt.plot(X_train,Y_train,"b.",label="Training Points")
plt.plot(X_test,Y_test,"g.",label="Testing Points")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```
