```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("student_scores.csv")
```

```python
df.head(10)
```

```python
plt.scatter(df['Hours'],df['Scores'],color = "green",marker='*',s= 100)
plt.xlabel("Study_Hours")
plt.ylabel("Score_IN_Exam")
```

```python
# First find Correlation
df.corr()
```

```python
# Now visualize the data
sns.pairplot(df)
```

```python
# identify independent and dependent feature
X = df[["Hours"]]
Y = df["Scores"]
```

```python
# Now Train and Test split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size = 0.25,random_state=42)
```

```python
X_train.shape
```

```python
# Now Do Standardization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
X_test
```

```python
# Now apply Linear Regression
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
regression.fit(X_train,Y_train)
```

```python
print("Coefficient is = ",regression.coef_)# it's mean that one unit moment in the axis the value of y will be 22.28 
print("Intercept : ",regression.intercept_) # and it's mean that when the value of x is zero then the intercept is 156.47 (meet to y point is 61.625)
```

```python
plt.scatter(X_train,Y_train)
plt.plot(X_train,regression.predict(X_train)) # best fit line
```

```python
# Now for new Data
newStudyHour = float(input("Enter Your Study Hours To Predict Your Score = "))
predictScore = regression.predict(scaler.transform([[newStudyHour]]))
print("Your Study Score will be = ",predictScore)
```
