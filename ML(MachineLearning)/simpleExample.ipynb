```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
reg = LinearRegression()
```

```python
X = [[1],[2],[3],[4],[5],[6]]
Y = [2,2.5,4.5,3,5,4.7]

reg.fit(X,Y)
```

```python
plt.scatter(X,Y)
plt.plot(X,reg.predict(X))
plt.show()
newPoint = 3.3
predictedPoint = reg.predict([[newPoint]])
plt.plot(X,reg.predict(X))
plt.scatter(newPoint,predictedPoint,color='green', marker='*', s=100)
```
